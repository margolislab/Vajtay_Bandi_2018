```python
%matplotlib inline
from brokenaxes import brokenaxes
import matplotlib.pyplot as plt
import numpy as np
import scipy.io as sci
from matplotlib.ticker import AutoMinorLocator

minor_locator = AutoMinorLocator(2)
minor_locatorx = AutoMinorLocator(2)

plt.rcParams['font.family'] = "Arial"
plt.rcParams.update({'font.size': 10})
plt.rcParams['savefig.bbox'] = 'tight'

mat = sci.loadmat('chat-M4trace.mat')

bar_input = mat['bar_input']

x = np.arange(0,3000,2)

f, ax = plt.subplots(1, 4, figsize=(8.5,3))
f.subplots_adjust(wspace = 0.4)

n = 0
for j in range(0, len(ax)):
    if j < 3:
        ax[j].plot(x, bar_input[0:1500,3:7,j], '0.6',alpha = 0.5)
        ax[j].plot(x, bar_input[0:1500,0:2,j], '0.6', alpha = 0.5)
        ax[j].plot(x,bar_input[0:1500,7,j], 'k')
        ax[j].plot(x,bar_input[0:1500,2,j], 'r', alpha = 0.7)
        ax[j].set_ylim(-3,15)
        ax[j].set_xlim(700,2500)
        ax[j].spines['top'].set_visible(False)
        ax[j].spines['right'].set_visible(False)
        ax[j].spines['bottom'].set_linewidth(0.5)
        ax[j].spines['left'].set_linewidth(0.5)
    if j > 0:
        ax[j].spines['left'].set_visible(False)
        ax[j].yaxis.set_visible(False)
    if j == 3:
        ax[j].plot(x, bar_input[0:1500,0,j], '0.6')
        ax[j].set_ylim(-3,15)
        ax[j].set_xlim(700,2500)
        ax[j].spines['top'].set_visible(False)
        ax[j].spines['right'].set_visible(False)
        ax[j].spines['bottom'].set_linewidth(0.5)
        ax[j].spines['left'].set_linewidth(0.5)
    
ax[0].set_ylabel('Angle (\xb0)')
ax[0].set_xlabel('Time (ms)')

plt.show()
```

```python
mat = sci.loadmat('chat_av.mat')

bar_input = mat['bar_input']

x = [-1,0,1,2]
y1 = bar_input[:,2,0]
error1 = bar_input[:,2,1]
y2 = bar_input[:,7,0]
error2 = bar_input[:,7,1]

ax = plt.subplot()
ax.plot(x, y1,color='r')
ax.plot(x, y1,'ro')
plt.fill_between(x, y1-error1, y1+error1,facecolor='0.7')
ax.plot(x, y2,color='k')
ax.plot(x, y2,'ko')
ax.vlines(0,-1,15,color = 'k', linestyles = 'dashed')
plt.fill_between(x, y2-error2, y2+error2,facecolor='0.7')
ax.set_ylabel('Peak angle (\xb0)')
ax.set_xlabel('Hours after lesion')

plt.xticks([-1,0,1,2],('Pre','0.5','6','24'))
plt.yticks([0,5,10,15,20,25,30])
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.yaxis.set_minor_locator(minor_locator)
ax.tick_params(which='major', length=5)
ax.tick_params(which='minor', length=3)
ax.set_ylim(-1,15)
fig = plt.gcf()
fig.set_size_inches(2,2)

plt.show()
```

```python
mat = sci.loadmat('chat_ttp.mat')
mat2 = sci.loadmat('chat_hw.mat')
bar_input = mat['bar_input']
bar_input2 = mat2['bar_input']

x = [-1,0,1,2]
y1 = bar_input[0:3,7,0]
error1 = bar_input[0:3,7,1]
y2 = bar_input2[0:3,7,0]
error2 = bar_input2[0:3,7,1]
y1 = np.append(y1, 0)
y2 = np.append(y2, 0)
error1 = np.append(error1, 0)
error2 = np.append(error2, 0)


ax = plt.subplot()
ax.plot(x, y1,color='k')
ax.plot(x, y1,'k^', label = 'Time-to-peak')
plt.fill_between(x, y1-error1, y1+error1,facecolor='0.7')
ax.plot(x, y2,color='k')
ax.plot(x, y2,'kx', label = 'Half-width')
ax.vlines(0,-4,31,color = 'k', linestyles = 'dashed')
plt.fill_between(x, y2-error2, y2+error2,facecolor='0.7')
ax.set_ylabel('Time-to-peak, Half-width (ms)')
ax.set_xlabel('Hours after lesion')
plt.xticks([-1,0,1,2],('Pre','0.5','6','24'))
plt.yticks([0,10,20,30])
ax.yaxis.set_minor_locator(minor_locator)
plt.tick_params(which='major', length=5)
plt.tick_params(which='minor', length=3)
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.set_ylim(-3,30)
ax.legend(fontsize = 9)

fig = plt.gcf()
fig.set_size_inches(2.5,2.5)

plt.show()
```

```python
mat = sci.loadmat('chat_av.mat')

bar_input = mat['bar_input']
x = [1,3,5,10,20,50,100,1000]
y1 = bar_input[0,:,0]
error1 = bar_input[0,:,1]
y2 = bar_input[2,:,0]
error2 = bar_input[2,:,1]

bax = brokenaxes(xlims=((-10, 109),(990, 1005)), wspace = 0.1)

bax.set_ylim(-1,12)
bax.plot(x, y1, '0.4')
bax.plot(x, y2, '0.4')
bax.plot(x[0:2],y1[0:2],'o',color = '0.4')
bax.plot(x[3:7],y1[3:7],'o',color = '0.4', label='Pre-lesion')
bax.plot(x[0:2],y2[0:2],'o',color = '0.4')
bax.plot(x[3:7],y2[3:7],'o',color = '0.4', label='+6 Hours')
bax.plot(x[2],y1[2],'ro')
bax.plot(x[2],y2[2],'ro')
bax.plot(x[7],y1[7],'ko')
bax.plot(x[7],y2[7],'ko')
bax.fill_between(x, y1-error1, y1+error1,facecolor='0.7')
bax.fill_between(x, y2-error2, y2+error2,facecolor='0.7')
bax.set_ylabel('Peak angle (\xb0)', labelpad = 25)
bax.set_xlabel('Stimulus duration (ms)', labelpad = 25)
bax.axs[0].yaxis.set_minor_locator(minor_locator)
bax.axs[0].xaxis.set_minor_locator(minor_locatorx)
bax.tick_params(which='major', length=5)
bax.tick_params(which='minor', length=3)
bax.axs[0].set_yticks([0,4,8,12])
bax.axs[0].set_xticks([0,50,100])

fig = plt.gcf()
fig.set_size_inches(2,2)


plt.show()
```
