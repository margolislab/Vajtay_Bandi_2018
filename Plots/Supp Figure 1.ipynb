```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import scipy.io as sci

plt.rcParams['font.family'] = "Arial"
plt.rcParams.update({'font.size': 10})
plt.rcParams['savefig.bbox'] = 'tight'

mat = sci.loadmat('10Hz.mat')

bar_input = mat['bar_input']

x = np.arange(0,3000,2)

f, ax = plt.subplots(2, 2)
f.subplots_adjust(wspace = 0.4)

x1 = np.arange(3000)
y1 = np.repeat(-6,1000)
y2 = []

for j in range(10):
    stim = np.repeat(-3,5)
    pause = np.repeat(-6,95)
    y2 = np.concatenate((y2,stim,pause))

y10Hz = np.concatenate((y1,y2,y1))

y3 = np.repeat(-3,1000)

y1000ms = np.concatenate((y1,y3,y1))

ystim = np.array([y10Hz, y1000ms])


a = 0

for i in range(2):
    for j in range(2):
        ax[i,j].plot(x[400:1100],bar_input[400:1100,a], 'k')
        ax[i,j].plot(x1[800:2200],ystim[i,800:2200],'cyan',linewidth=2)
        ax[i,j].set_ylim(-6,25)
        ax[i,j].spines['top'].set_visible(False)
        ax[i,j].spines['right'].set_visible(False)
        #ax[i,j].spines['left'].set_visible(False)
        ax[i,j].spines['bottom'].set_visible(False)
        ax[i,j].set_xticks([])
        ax[i,j].set_yticks([0,10,20])
        a += 1
    
ax[1,0].set_ylabel('Angle(\xb0)')
#ax[1,0].set_xlabel('Time(ms)')

ax[0,0].set_title('Pre-lesion')
ax[0,1].set_title('Day +7')

f.set_size_inches(7,5)

f.savefig('EMX_10Hz_M2.svg')

plt.show()
```
