```python
%matplotlib inline
from brokenaxes import brokenaxes
import matplotlib.pyplot as plt
import numpy as np
import scipy.io as sci
from matplotlib.ticker import AutoMinorLocator

minor_locator = AutoMinorLocator(2)
minor_locatorx = AutoMinorLocator(2)

plt.rcParams['font.family'] = "Arial"
plt.rcParams.update({'font.size': 10})
plt.rcParams['savefig.bbox'] = 'tight'

mat = sci.loadmat('emx-M2trace_r.mat')

bar_input = mat['bar_input']

x = np.arange(0,3000,2)

f, ax = plt.subplots(1, 4, figsize=(8.5,3))
f.subplots_adjust(wspace = 0.4)

for j in range(0, len(ax)):
    ax[j].plot(x, bar_input[0:1500,3:7,j], '0.6',alpha = 0.5)
    ax[j].plot(x, bar_input[0:1500,0:2,j], '0.6', alpha = 0.5)
    ax[j].plot(x,bar_input[0:1500,7,j], 'k')
    ax[j].plot(x,bar_input[0:1500,2,j], 'r', alpha = 0.7)
    ax[j].set_ylim(-20,10)
    ax[j].set_xlim(700,2500)
    ax[j].spines['top'].set_visible(False)
    ax[j].spines['right'].set_visible(False)
    ax[j].spines['bottom'].set_linewidth(0.5)
    ax[j].spines['left'].set_linewidth(0.5)
    if j > 0:
        ax[j].spines['left'].set_visible(False)
        ax[j].yaxis.set_visible(False)
    
ax[0].set_ylabel('Angle(\xb0)')
ax[0].set_xlabel('Time(ms)')

plt.show()
```

```python
mat = sci.loadmat('emx-avr.mat')

bar_input = mat['bar_input']
ebars = mat['ebars']
x = [-2,0,1,2,3,4,7,8,9,10]
y1 = bar_input[:,2]
error1 = ebars[:,2]
y2 = bar_input[:,7]
error2 = ebars[:,7]


fig, bax = plt.subplots()
bax.plot(x, y1,'ro-')
bax.fill_between(x, y1-error1, y1+error1,facecolor='0.7')
bax.plot(x, y2,'ko-')
bax.fill_between(x, y2-error2, y2+error2,facecolor='0.7')
bax.vlines(0,-20,3,color = 'k', linestyles = 'dashed')
bax.set_ylim(-20,3)
bax.set_xlim(-3,-1)
plt.xticks([-2,0,2,4,6,8,10],('Pre','0','2','4','6','8','10'))
bax.set_yticks([-20,-10,0])
bax.yaxis.set_minor_locator(minor_locator)
bax.xaxis.set_minor_locator(minor_locatorx)
plt.tick_params(which='major', length=5)
plt.tick_params(which='minor', length=3)
bax.spines['top'].set_visible(False)
bax.spines['right'].set_visible(False)
bax.set_ylabel('Peak angle (\xb0)')
bax.set_xlabel('Days after lesion')
bax.set_ylim(-20,3)
fig = plt.gcf()
fig.set_size_inches(2,2)

#ax.set_title('EMX 5,1000ms Stimuli Over Days')
plt.show()
```

```python
mat = sci.loadmat('emx-av_vr.mat')

bar_input = mat['bar_input']
ebars = mat['ebars']
x = [1,3,5,10,20,50,100,1000]
y1 = bar_input[0,:]
error1 = ebars[0,:]
y2 = bar_input[6,:]
error2 = ebars[6,:]

bax = brokenaxes(xlims=((-10, 109),(990, 1005)), wspace = 0.1)
bax.set_ylim(-1000,100)
bax.plot(x, y1, '0.4')
bax.plot(x, y2, '0.4')
bax.plot(x[0:2],y1[0:2],'o',color = '0.4')
bax.plot(x[3:7],y1[3:7],'o',color = '0.4', label='Baseline')
bax.plot(x[0:2],y2[0:2],'o',color = '0.4')
bax.plot(x[3:7],y2[3:7],'o',color = '0.4', label='+7 Days')

bax.plot(x[2],y1[2],'ro')
bax.plot(x[2],y2[2],'ro')
base = bax.plot(x[7],y1[7],'ko')
seven = bax.plot(x[7],y2[7],'ko')

bax.fill_between(x, y1-error1, y1+error1,facecolor='0.7')
bax.fill_between(x, y2-error2, y2+error2,facecolor='0.7')

bax.set_ylabel('Peak velocity (\xb0/s)', labelpad = 40)
bax.set_xlabel('Stimulus duration (ms)', labelpad = 25)

bax.axs[0].yaxis.set_minor_locator(minor_locator)
bax.axs[0].xaxis.set_minor_locator(minor_locatorx)
bax.tick_params(which='major', length=5)
bax.tick_params(which='minor', length=3)
bax.axs[0].set_yticks([-1000, -500, 0])
bax.axs[0].set_xticks([0,50,100])

fig = plt.gcf()

fig.set_size_inches(2,2)

plt.show()
```
